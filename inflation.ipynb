```python
import cpi
import datetime
import pandas as pd
cpi.update()
""" Calculate the real buy power of value by CPI Inflation Calculator. I plug the result into real_dollars column
    """
"""take SD as an example"""
def graph_inflation_comparison(county_name):
    
    datal = data.loc[data['County'] == county_name,['date','value']]
    for i in datal['date']:
        change_date(i) 
    datal["year"] = pd.DatetimeIndex(datal['date']).year
    # datal["year"] = data.index.year
    d1=[int(i) for i in list(data1['value'].values)]
    cpi.update()
    datal["real_dollars"] = datal.apply(lambda x: cpi.inflate(x["value"], x["year"]), axis=1)
    c = (
        Line(init_opts=opts.InitOpts(theme=ThemeType.DARK)))
    dl = [int(j) for j in  list(datal['value'].values)]
    d2 = [int(j) for j in  list(datal['real_dollars'].values)]
    y=[str(k) for k in list(datal['date'].values)]
    c.add_xaxis(y)
    c.add_yaxis("pre_process_value",dl,is_symbol_show = False)
    c.add_yaxis("price_drop_inflation",d2,is_symbol_show = False)
    c.render_notebook()

```
